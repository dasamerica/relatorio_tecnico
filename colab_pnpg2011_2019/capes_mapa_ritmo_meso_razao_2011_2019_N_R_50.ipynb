```
# Mapa Doutorado Academico Matriculado
mapa_DA_M = folium.Map(
    width=800, height=600,
    location=[-16.1237611, -59.9219642], 
    tiles='http://tile.stamen.com/toner-lite/{z}/{x}/{y}.png ', 
    attr="toner-bcg",
    zoom_start=4
)
folium.GeoJson(geo_json_data).add_to(mapa_DA_M)

#Escala para discentes titulados 
#myscale = (DA_M['qty'].quantile(([0, 0.01, 0.02, 0.03, 0.07, 0.20, 0.30, 0.50, 1]))).tolist()
myscale =[-100.0, -50.0, -25, -10, 0.0001, 10.0, 25.0, 50.0, 100.0]

lc = cm.linear.YlGnBu_06
lc.to_step(data=myscale, n=12, method='linear')


folium.Choropleth(
    geo_data=geo_json_data,
    name='Doutorado Academico Matriculado',
    data=DA_M,
    columns=['id','qty'],
    key_on='feature.properties.MESO',
 #   fill_color='YlOrRd',
    fill_color='RdBu',
    threshold_scale=myscale,
    control_scale=True,
    fill_opacity=1,
    line_opacity=0.1,
    legend_name='% alteracao quantidade de discentes por habitantes - media a.a.',
    nan_fill_color = 'gray',
    nan_fill_opacity=0.02
).add_to(mapa_DA_M)

style_function = lambda x: {'fillColor': 'green', 
                            'color':'green', 
                            'fillOpacity': 0.50, 
                            'weight': 0.1
                            }
highlight_function = lambda x: {'fillColor': 'green', 
                                'color':'green', 
                                'fillOpacity': 0.10, 
                               'weight': 0.1
                                }
NIL = folium.features.GeoJson(
    nil_DA_M_N,
    style_function=style_function, 
    control=False,
    highlight_function=highlight_function
)

mapa_DA_M.add_child(NIL)
mapa_DA_M.keep_in_front(NIL)

folium.LayerControl().add_to(mapa_DA_M)
mapa_DA_M.save('sample_data/index_dam.html')
```
